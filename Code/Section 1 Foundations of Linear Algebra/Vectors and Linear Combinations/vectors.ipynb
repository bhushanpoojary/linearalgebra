```python
# Python Code to Show Vectors as Arrows

import matplotlib.pyplot as plt

# Define vectors
vectors = [(2, 1), (1, 3)]

# Create figure
fig, ax = plt.subplots()
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

# Plot each vector from origin (0,0)
for v in vectors:
    ax.arrow(0, 0, v[0], v[1], head_width=0.2, head_length=0.3, fc='blue', ec='black')

# Add grid and labels
ax.grid(True)
ax.set_aspect('equal')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Vectors from Origin')

plt.show()
```

```python
#Tail-to-Tip Vector Addition
import matplotlib.pyplot as plt

# Define vectors
v1 = [1, 2]
v2 = [3, 1]
v3 = [v1[0] + v2[0], v1[1] + v2[1]]  # Resultant vector v1 + v2

# Create plot
fig, ax = plt.subplots()
ax.set_xlim(0, 5)
ax.set_ylim(0, 4)

# Plot v1 from origin
ax.arrow(0, 0, v1[0], v1[1],
         head_width=0.1, head_length=0.2,
         fc='blue', ec='blue')

# Plot v2 from tip of v1 (tail-to-tip)
ax.arrow(v1[0], v1[1], v2[0], v2[1],
         head_width=0.1, head_length=0.2,
         fc='green', ec='green')

# Plot v2 from origin
ax.arrow(0, 0, v2[0], v2[1],
         head_width=0.1, head_length=0.2,
         fc='green', ec='green')

# Plot resultant vector v1 + v2 from origin
ax.arrow(0, 0, v3[0], v3[1],
         head_width=0.1, head_length=0.2,
         fc='red', ec='red', linestyle='dashed')

# Add labels
ax.text(0.3, 1, 'v1', fontsize=12, color='blue')
ax.text(3, 2.3, 'v2', fontsize=12, color='green')
ax.text(2.5, 1.2, 'v1 + v2', fontsize=12, color='red')

# Grid and aspect
ax.grid(True)
ax.set_aspect('equal')
plt.title('Vector Addition: Tail-to-Tip Method')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#Scalar Multiplication in Python
import matplotlib.pyplot as plt

# Define the original vector
v = [1, 3]
scalar = 2

# Resulting vector after scalar multiplication
v_scaled = [scalar * v[0], scalar * v[1]]

# Create plot
fig, ax = plt.subplots()
ax.set_xlim(0, 4)
ax.set_ylim(0, 7)

# Plot original vector
ax.arrow(0, 0, v[0], v[1],
         head_width=0.1, head_length=0.2,
         fc='blue', ec='blue', label='v')

# Plot scaled vector
ax.arrow(0, 0, v_scaled[0], v_scaled[1],
         head_width=0.1, head_length=0.2,
         fc='green', ec='green', label='2·v')

# Add labels
ax.text(0.5, 1.5, 'v = [1, 3]', color='blue', fontsize=12)
ax.text(2, 5.5, '2·v = [2, 6]', color='green', fontsize=12)

# Decorate plot
ax.grid(True)
ax.set_aspect('equal')
plt.title('Scalar Multiplication of a Vector')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#Linear Combination Demo
import matplotlib.pyplot as plt

# Define unit vectors (basis vectors)
v1 = [1, 0]
v2 = [0, 1]

# Scalars
a = 3
b = 2

# Resulting vector: linear combination
result = [a * v1[0] + b * v2[0], a * v1[1] + b * v2[1]]

# Create plot
fig, ax = plt.subplots()
ax.set_xlim(0, 4)
ax.set_ylim(0, 3)

# Plot a * v1 from origin
ax.arrow(0, 0, a * v1[0], a * v1[1], head_width=0.1, head_length=0.2, fc='blue', ec='blue')
ax.text(1.2, 0.1, f'{a}·v1', color='blue')

# Plot b * v2 from tip of a*v1 (tail-to-tip)
ax.arrow(a * v1[0], a * v1[1], b * v2[0], b * v2[1], head_width=0.1, head_length=0.2, fc='green', ec='green')
ax.text(3.1, 1, f'{b}·v2', color='green')

# Plot result from origin
ax.arrow(0, 0, result[0], result[1], head_width=0.1, head_length=0.2, fc='red', ec='red', linestyle='dashed')
ax.text(2.3, 1.6, f'{a}·v1 + {b}·v2', color='red')

# Decorations
ax.grid(True)
ax.set_aspect('equal')
plt.title('Linear Combination: 3·v1 + 2·v2')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
